```python
#Feature Transformation - PowerTransformation
from numpy import log1p
from sklearn.preprocessing import PowerTransformer
x=df[['ApplicantIncome','CoapplicantIncome']].dropna()
pt=PowerTransformer(method='yeo-johnson')
x_power_transformed=pd.DataFrame(pt.fit_transform(x),columns=['ApplicantIncome','CoapplicantIncome'])
fig,axes=plt.subplots(2,2,figsize=(12,10))
sns.histplot(x['ApplicantIncome'],bins=30,color='yellow',kde=True,ax=axes[0,0])
axes[0,0].set_title('Original Data of ApplicantIncome')
sns.histplot(x_power_transformed['ApplicantIncome'],bins=30,kde=True,color='yellow',ax=axes[0,1])
axes[0,1].set_title('Power Transformed Data off ApplicantIncome')
sns.histplot(x['CoapplicantIncome'],bins=30,kde=True,color='orange',ax=axes[1,0])
axes[1,0].set_title('Original Data of CoapplicantINcome')
sns.histplot(x_power_transformed['CoapplicantIncome'],bins=30,kde=True,color='orange',ax=axes[1,1])
axes[1,1].set_title('Power Transformed Data of CoapplicantIncome')
```
